### Task:4
#### Find the goal diff count, team that has max and min goal diff in the same function and return it.

```python
'''
As the request is not to add a create column in the data frame, 
Used .assign() rather than .insert() as it adds the col in the same DF
'''
def Goal_diff_count():
    
    # a new data frame
    ds_view = DS.assign( Goal_diff_cnt = DS['GoalsFor'] - DS['GoalsAgainst'])
    
    # As the diff results in negative value, used abs()
    ds_view['Goal_diff_cnt']=ds_view['Goal_diff_cnt'].abs()   
    
    # Find the index that has max and min value
    idx_min = ds_view['Goal_diff_cnt'].idxmin()
    idx_max = ds_view['Goal_diff_cnt'].idxmax()
    
    # return the data fram, max/min value for further references
    return ds_view,idx_min,idx_max
```

```python
# Function calling and assigning the return values
ds_view, idx_min, idx_max = Goal_diff_count()
```

```python
ds_view.iloc[idx_min,]
```

```python
ds_view.iloc[idx_max,]
```
